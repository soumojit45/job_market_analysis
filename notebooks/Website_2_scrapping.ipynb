# 1. At first collect all 254 pages

```python
webPageList = []    
# list to store all html pages 
for i in range(1,255):
    WebText = requests.get(f'https://internshala.com/jobs/page-{i}/',headers=headers).text
    soup = BeautifulSoup(WebText,'html.parser')
    webPageList.append(soup)
```

# 2. page by page , then card by card data extraction

```python
def extract_salary_float(salary_text):
        if not salary_text:
            return np.nan
        try:
            salary = salary_text.strip()[2:]
            sal_list = salary.split("-")
            ans = [i.strip().replace(',','') for i in sal_list]
            sal_list2 = [float(i) for i in ans]
            return sum(sal_list2) / len(sal_list2)
        except ValueError:
            return np.nan

all_jobs_data = []

for soup in webPageList:
    
    job_cards = soup.find_all('div', class_='individual_internship')
    
    for card in job_cards:
        #Company Name
        company_tag = card.find('p', class_='company-name')
        company_name = company_tag.text.strip() if company_tag else np.nan
        
        #Job Post
        post_tag = card.find('h2', class_='job-internship-name')
        job_post = post_tag.text.strip() if post_tag else np.nan
        
        #Location
        location_tag = card.find('p', class_='locations')
        if location_tag:
            # Extract all anchor tags <a> inside the location paragraph
            locations = [a.text.strip() for a in location_tag.find_all('a')]
            location = ", ".join(locations) if locations else np.nan
        else:
            location = np.nan
            
        #Job Salary
        job_salary = np.nan
        money_icon = card.find('i', class_='ic-16-money')
        if money_icon and money_icon.parent:
            salary_span = money_icon.parent.find('span', class_='desktop')
            if salary_span:
                job_salary = extract_salary_float(salary_span.text.strip())
                
        # --- Experience Needed ---
        experience_needed = np.nan
        briefcase_icon = card.find('i', class_='ic-16-briefcase')
        if briefcase_icon and briefcase_icon.parent:
            exp_span = briefcase_icon.parent.find('span')
            if exp_span:
                experience_needed = exp_span.text.strip()
                
        # Append row data
        all_jobs_data.append({
            'Company Name': company_name,
            'Job Post': job_post,
            'Location': location,
            'Job Salary': job_salary,
            'Experience Needed': experience_needed
        })
```

# 3. Convert extracted data into a Pandas DataFrame
