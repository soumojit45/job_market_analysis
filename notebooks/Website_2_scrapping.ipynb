```python
WebText = requests.get('https://internshala.com/jobs/page-1/',headers=headers).text
soup = BeautifulSoup(WebText,'html.parser')
print(soup)
```

# fetch company name

```python
for i in range(10,15):
    try:
        print(soup.find_all('p',class_="company-name")[i].text.strip())
    except IndexError:
        print(i)
        break
```

# fetch job post 

```python
for i in range(10,15):
    print(soup.find_all('a',class_="job-title-href")[i].text.strip())
```

# fetch job loaction

```python
for i in soup.find_all('p', class_='row-1-item locations'):
    if(i):
        location_tags = i.find_all('a')
        locations = [tag.text.strip() for tag in location_tags]
        print(locations)
```
